```python
from tqdm import tqdm
tqdm.pandas()
```

```python
data = pd.read_csv("final_dataset.csv")
data.head()
```

```python
# Spliting words into two meaningful word

import wordninja as wn

data["sentence"] = data.sentence.progress_apply(lambda x: ' '.join(wn.split(str(x))))
```

```python
data.drop(columns=["tokens", "lemmatizer", "sentence_cleaned"], inplace=True)
```

```python
data.shape
```

```python
import nltk
# nltk.download('punkt')

data['tokens'] = data.sentence.progress_apply(lambda x: nltk.word_tokenize(x))
data
```

```python
from nltk.stem import WordNetLemmatizer
# nltk.download('wordnet')
# nltk.download('omw-1.4')

lm = WordNetLemmatizer()

data['lemmatizer'] = data.tokens.progress_apply(lambda x: [lm.lemmatize(word) for word in x])
data
```

```python
data.to_csv("latest.csv", index=False)
```
